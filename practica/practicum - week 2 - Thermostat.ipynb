# Implementing the Thermostat Game

Now, let's implement the physical system.

Version 1: Temperature changes.


```r
deltaTemp = 0.5

Temps = c(6)
for (tt in 1:60) {
    Temps = c(Temps, tail(Temps, 1) + deltaTemp)
}
```

```r
plot(Temps)
```

Version 2: Simple negative feedback



```r
goalTemp = 4

Temps = c(6)
for (tt in 1:60) {
    Temps = c(Temps, tail(Temps, 1) + goalTemp - tail(Temps, 1))
}
```

```r
plot(Temps)
```

Version 2: Simple negative feedback



```r
goalTemp = 4
rate = 0.5

Temps = c(6)
for (tt in 1:60) {
    Temps = c(Temps, tail(Temps, 1) + rate * (goalTemp - tail(Temps, 1)))
}
```

```r
plot(Temps)
```

Version 3: Add in a delay



```r
goalTemp = 4
rate = 0.5

Temps = c(6, 6, 6)
for (tt in 1:60) {
    Temps = c(Temps, tail(Temps, 1) + rate * (goalTemp - Temps[length(Temps) - 2]))
}
```

```r
plot(Temps)
```

Version 4: Add in the human



```r
goalTemp = 4
rate = 0.5

Temps = c(6, 6, 6)
for (tt in 1:60) {
    anticipatoryGoalTemp = goalTemp + (goalTemp - Temps[length(Temps)])
    Temps = c(Temps, tail(Temps, 1) + rate * (anticipatoryGoalTemp - Temps[length(Temps) - 2]))
}
```

```r
plot(Temps)
```

Version 5: Human anxiety



```r
plot(Temps)
```

```r
goalTemp = 4
rate = 0.5

Temps = c(6, 6, 6)
Overheat = c(2)
for (tt in 1:60) {
    Overheat = c(Overheat, tail(Overheat, 1) + tail(Temps, 1) - goalTemp)
    anticipatoryGoalTemp = goalTemp - tail(Overheat, 1)
    Temps = c(Temps, tail(Temps, 1) + rate * (anticipatoryGoalTemp - Temps[length(Temps) - 2]))
}
```

```r
plot(Temps, type='l')
```
