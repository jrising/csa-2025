```r
rate = 0.1
Temps = c(6)

for (tt in 1:60) {
    Temps = c(Temps, Temps[tt] + rate)
}
```

```r
plot(Temps)
```

```r
rate = 0.1
goal = 4
Temps = c(6)

for (tt in 1:60) {
    Temps = c(Temps, Temps[tt] + goal - Temps[tt])
}
```

```r
plot(Temps)
```

```r
rate = 0.1
goal = 4
Temps = c(6)

for (tt in 1:60) {
    Temps = c(Temps, Temps[tt] + rate * (goal - Temps[tt]))
}
```

```r
plot(Temps)
```

```r
rate = 0.1
goal = 4
Temps = c(6, 6, 6)

for (tt in 3:60) {
    Temps = c(Temps, Temps[tt] + rate * (goal - Temps[tt - 2]))
}
```

```r
plot(Temps, type='l')
```
